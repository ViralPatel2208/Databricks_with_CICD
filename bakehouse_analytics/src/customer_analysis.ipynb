```python
# Set default catalog and schema
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {schema}")
spark.sql(f"USE SCHEMA {schema}")
```

```python
spark.read.table("samples.nyctaxi.trips")
```

```python
spark.sql(f"""
          create or replace table {catalog}.{schema}.customer_analysis as
          select country,
          COUNT(*) AS NUMBER_OF_CUSTOMER,
          COUNT_IF(GENDER='MALE') AS MALE_CUSTOMER,
          COUNT_IF(GENDER='FEMALE') AS FEMALE_CUSTOMER
          FROM samples.bakehouse.sales_customers
          group by country""");
```
